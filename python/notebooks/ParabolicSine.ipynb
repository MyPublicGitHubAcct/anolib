```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constants
B = 4.0 / np.pi
C = -4.0 / (np.pi * np.pi)
P = 0.225
t = np.linspace(-np.pi, np.pi, 256) # array from -pi to pi
```

```python
def unipolarToBipolar(value):
    """ convert 0 to 1 to -1 to 1 """
    return 2.0*value - 1.0

def bipolarToUnipolar(value):
    """ convert -1 to 1 to 0 to 1 """
    return 0.5*value + 0.5

def parabolic_sine(x):
    """
    :param x: A value between -pi and pi
    :return: The next value of a sine waveform
    """
    y = B * x + C * x * abs(x)
    return P * (y * abs(y) - y) + y
```

```python
out1 = parabolic_sine(t)
fig1 = plt.figure("Figure 1")
plt.plot(out1)
plt.show()
```
