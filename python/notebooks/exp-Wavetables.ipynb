```python
%matplotlib ipympl
import numpy as np
# from scipy.signal import sawtooth
import matplotlib.pyplot as plt
```

```python
WAVETABLE_LENGTH = 128
```

```python
sineWaveTable = []
for i in range(WAVETABLE_LENGTH):
    sineWaveTable.append(np.sin(2 * np.pi * i / WAVETABLE_LENGTH))
```

```python
# print(sineWaveTable)
plt.plot(sineWaveTable)
plt.show()
```

```python
def first_order_linear_interpolator(i, k, wt):
    """
    :param i: current index
    :param k: wave-table index
    :param wt: wave-table array
    :return: The weighted sum of wave table values at
             curr_idx and curr_idx+1.
    """
    return (k - i) * wt[i + 1] + (i + 1 - k) * wt[i]
```

```python
first_order_linear_interpolator(10, 10, sineWaveTable)
```
