# Plotting an FFT

Based on _Signal Processing with Paul_'s __Plotting the Fourier Transform in Python (DFT/FFT)__ - https://youtu.be/arYKpiVACvY

```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define x(t) as a pulse function
num_t = 100
xt = np.zeros(num_t)
# xt[:10] = 1
xt[:5] = 1
xt[-4:] = 1

fig1 = plt.figure("Waveform")
plt.plot(xt)
plt.title("Plot of x(t)")
plt.xlabel("Sample index")
plt.ylabel("Value of x(t)")
plt.show()
```

## Plotting the FFT




```python
# Do the fft and plot it
fxt = np.fft.fft(xt)
# First plot is the real part
fig2 = plt.figure("FFT")
plt.subplot(1, 2, 1)
plt.plot(np.fft.fftshift(np.real(fxt)))
plt.title("Plot of real part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
# Second plot is the imaginary part
plt.subplot(1, 2, 2)
plt.plot(np.fft.fftshift(np.imag(fxt)))
plt.title("Plot of imaginary part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
plt.show()
```

```python
# Plot in polar form
fig3 = plt.figure("Polar form")
# First plot is the real part
plt.subplot(1, 2, 1)
plt.plot(np.fft.fftshift(np.abs(fxt)))
plt.title("Magnitude of the FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
# Second plot is the imaginary part
plt.subplot(1, 2, 2)
plt.plot(np.fft.fftshift(np.angle(fxt)))
plt.title("Angle of the FT of x(t)")
plt.ylabel("Angle")
plt.xlabel("Frequency Index")
plt.show()
```
