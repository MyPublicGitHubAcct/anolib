Based on the wavetable oscillator described by https://www.earlevel.com/ 

```python
%matplotlib ipympl
import numpy as np
from scipy import signal
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
length = 2048
t = np.linspace(0, 1, length)
saw_wave = signal.sawtooth(2 * np.pi * 1 * t)
```

```python
# Number of sample points
N = length
# sample spacing
T = 1.0 / length


yf = fft(saw_wave)
# xf = np.linspace(0.0, 1.0 / (2.0*T), N//2)
# plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))

plt.plot(t[0:length//2], 2.0/length * np.abs(saw_wave[0:length//2]))


plt.show() 
```
